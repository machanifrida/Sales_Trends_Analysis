```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load sales data
df = pd.read_csv("sales_data.csv")
```

```python
df.head(5)
```

```python
#load customer data
df2 = pd.read_csv("customer_data.csv")
df2.head(5)
```

```python
#merge the datasets
data = pd.merge(df, df2, on = "customer_id")
data.sample(10)
```

```python
# Visualize sales trends
plt.figure(figsize=(12,6))
data.groupby('invoice date')['price'].sum().plot()
plt.title('Daily Sales Trend')
plt.show()
```

# Sales Trend Analysis

Analyze trends over time

```python
# Convert date to datetime
data['invoice date'] = pd.to_datetime(data['invoice date'])

# Extract month name
#data['month'] = data['invoice date'].dt.month_name()

# Extract time-based features
data['day_of_week'] = data['invoice date'].dt.day_name()
data['month'] = data['invoice date'].dt.month_name() # Extract month name
data['year'] = data['invoice date'].dt.year
```

```python
#calculate total sales
data["sales"] = data["price"] * data["quantity"]
```

```python
data.sample(10)
```

## Analyze trends over time
Aggregate data by time periods (e.g., days, weeks, or months) to analyze trends.

## Monthly trends

```python
#Group by month

monthly_sales = data.groupby("month")["sales"].sum().reset_index()
monthly_sales
```

```python
 # Convert period to string for plotting
monthly_sales['month'] = monthly_sales['month'].astype(str) 
```

```python
#Sort months by calendar order using pd.categorical function

month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

# Convert month column to categorical type
monthly_sales['month'] = pd.Categorical(monthly_sales['month'], categories=month_order, ordered=True)

# Sort by calendar order
monthly_sales = monthly_sales.sort_values('month')
```

### Visualize Sales Trends
Plot the sales trends over time using a line chart.

```python
plt.figure(figsize = (12, 6))
ax = plt.plot(monthly_sales['month'], monthly_sales['sales'], marker='o', color = "olive")
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

**Sales were highest in January and February and the lowest in November** 

**Sales hit the bottom in September, peaked up in October and hit the bottom again in November**

**The sales were stable in the months of March through July**

# Analyze purchasing patterns over time

Group data by day, week, or specific hours to identify peak purchasing times.

## Daily sales trends

```python
# Analyze total sales by day of the week
daily_sales = data.groupby("day_of_week")["sales"].sum().reset_index()
daily_sales["day_of_week"] = daily_sales["day_of_week"].astype(str)
```

```python
days_order = ["Sunday", "Monday", "Tuesday", "Thursday", "Friday", "Saturday"]
daily_sales["day_of_week"] = pd.Categorical(daily_sales["day_of_week"], categories = days_order, ordered = True)
```

```python
plt.figure(figsize = (12, 6))
sns.lineplot(daily_sales, x = "day_of_week", y = "sales", color = "brown")
plt.title("Daily Sales Trends", weight = "bold", size = 13)
plt.xlabel(" ")
plt.ylabel("Total Sales", size = 12)
plt.show()
```

**Sales peak on Mondays and are at lowest on Sundays**

# **Advanced Time Series Analysis**

### Smooth the trends using a rolling average to see long-term patterns.

```python
# Resample daily sales and compute a 7-day rolling average
daily_sales = data.groupby(data['invoice date'])['sales'].sum()
rolling_avg = daily_sales.rolling(window=7).mean()

plt.figure(figsize = (12, 6))
plt.plot(daily_sales, label='Daily Sales')
plt.plot(rolling_avg, label='7-Day Rolling Average', color='orange')
plt.title('Sales with Rolling Average')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

**The rolling average reflects a steady weekly trend even though the daily sales vary a lot**

## Seasonality and Trends

### Use time series decomposition to identify seasonality, trends, and noise.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(daily_sales, model='additive', period=30)

plt.figure(figsize = (12, 10))
#plt.rcParams['figure.figsize'] = [12, 8]
decomposition.plot()
plt.xticks(rotation = 45)
plt.show()
```

## **Conclusions and Further Actions**

The rolling average's narrower range suggests that sales are balanced over the week, with high-performing days compensating for weaker days.

The rolling average is not overly influenced by extreme values, indicating that high or low sales days are not disrupting overall trends.

Further actions should include: Planning inventory, staffing, and marketing efforts aligned with both the stable weekly trends and volatile daily patterns.

High-demand days such as Mondays, Fridays and Saturdays should be leveraged with promotions and inventory planning.

Low-demand days such as Sundays and Tuesdays signal opportunities for targeted marketing campaigns like discounts or special offers to boost sales during these slower periods.
